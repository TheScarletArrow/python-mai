```python
#import AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
import os
os.environ["OMP_NUM_THREADS"] = '1'
#import kmeans++
from sklearn.cluster import KMeans
#import matplotlib
import matplotlib.pyplot as plt
#set the environment variable OMP_NUM_THREADS=1
from sklearn import datasets
```

```python
#load iris dataset
iris = datasets.load_iris()
# take sepal length and petal width
iris_data = iris.data[:, [0, 3]]
```

```python
#split data into 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(iris_data)
#show the  data in clusters with different colors
plt.scatter(iris_data[y_kmeans == 0, 0], iris_data[y_kmeans == 0, 1], s=10, c='red', label='iris setosa')
#put legend in the upper left corner

plt.scatter(iris_data[y_kmeans == 1, 0], iris_data[y_kmeans == 1, 1], s=10, c='blue', label='iris versicolor')
plt.scatter(iris_data[y_kmeans == 2, 0], iris_data[y_kmeans == 2, 1], s=10, c='green', label='iris virginica')
plt.legend(loc='upper left')
```

```python
#learn iris dataset using AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3, linkage='complete')
y_agg = agg.fit_predict(iris_data)
#show the  data in clusters with different colors
plt.scatter(iris_data[y_agg == 0, 0], iris_data[y_agg == 0, 1], s=10, c='red', label='iris setosa')
plt.scatter(iris_data[y_agg == 1, 0], iris_data[y_agg == 1, 1], s=10, c='blue', label='iris versicolor')
plt.scatter(iris_data[y_agg == 2, 0], iris_data[y_agg == 2, 1], s=10, c='green', label='iris virginica')
plt.legend(loc='lower right')
```
